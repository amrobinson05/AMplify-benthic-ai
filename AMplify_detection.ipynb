```python
from ultralytics import YOLO

# Load a pretrained model (for transfer learning)
model = YOLO('yolo11m')

results = model.train(
    data='/content/drive/MyDrive/object_detection_dataset/data.yaml',
    epochs=100,           # Start high; early stopping will prevent overfitting
    imgsz=640,            # Standard size, good for most objects
    batch=16,             # Adjust based on GPU memory
    device='cuda',        # Use GPU
    augment=True,         # Essential for low object-per-image density
    val=True,             # Validate during training
    patience=10,          # Early stopping
    lr0=0.001,
    optimizer='AdamW',      # or 'AdamW' if you want faster convergence
    verbose=True
)
```
